## Population dynamics

- Note that the setting is different from the one used in paper, since we want to speed up the training. 
- One can set $m$ and $m_{target}$ larger, but the training will take more iterations.

```python
target = 'multi-neuron'
m = 20
d = 19
m_target = 5

nn = infinite.Train(m, d, target=target, m_target=m_target)
nn.run(nepochs=100000, learning_rate=1e-2, TOL=5e-6, plot_epoch=10000, check_epoch=50)
```

```python
plot_dynamics(nn, label='infinite')
```
